```python
print(type(ls_1923), len(ls_1923))
```

>**PREPROCESSING**

```python
#filtering text, removing weird characters
fil_1923 = re.sub('[^\\w\\s]', '', ls_1923)
print(type(fil_1923), len(fil_1923))
```

```python
#removing single letters try #1
pattern = r"((?<=^)|(?<= )).((?=$)|(?= ))"
filnew_1923 = re.sub("\s+", " ", re.sub(pattern, '', fil_1923).strip())
```

```python
print(type(filnew_1923), len(filnew_1923))
```

```python
#replace specific character
filnew_1923_no_g = filnew_1923.replace(' g ', '')
```

```python
#replace specific character
filnew_1923_no_g = filnew_1923_no_g.replace(' s ', '')
```

```python
print(type(filnew_1923_no_g), len(filnew_1923_no_g))
```

```python
#lower case the string
lower_1923 = str.lower(filnew_1923_no_g)
print(type(lower_1923), len(lower_1923))
```

```python
#create a pattern to remove digits
pattern = r"\d+"
# Match all digits in the string and replace them with an empty string
lower_1923_nodigits = re.sub(pattern, '', lower_1923)
print(len(lower_1923_nodigits), type(lower_1923_nodigits))
```

>**Tokenization**

```python
#tokenizing text into words
tokenized_1923 = nltk.word_tokenize(lower_1923_nodigits)
print(type(tokenized_1923))
print(len(tokenized_1923))
```

```python
print(tokenized_1923)
```

```python
#getting rid of no stops 
#eliminating no stops
nostops_1923 = [ t for t in tokenized_1923
                             if t not in stopwords.words('spanish')]
print(type(nostops_1923), len(nostops_1923))
```

```python
print(nostops_1923)
```

```python
#changing the format to fit the gensim dictionary
no_stops = [nostops_1923]
```

```python
print(no_stops)
```

>**GENSIM DICTIONARY**

```python
# Create a Dictionary from the articles: dictionary
dct = Dictionary(no_stops)
```

```python
#token to id system
#dct.token2id
```

```python
#checking lengths and types
print(len(dct), type(dct))
```

```python
print(dct)
```

>**gensim corpus**

```python
# Create a MmCorpus: corpus from the dictionary
corpus = [dct.doc2bow(doc) for doc in no_stops]
```

```python
corpus
```

```python
# Select the id for "computer": computer_id
box_id = dct.token2id.get("box")
print(box_id)
```

```python
# Print the first 10 word ids with their frequency counts from the fifth document
print(corpus[0][:10])
```

```python
#picking one of the list within the list
doc = corpus[0]
```

```python
# Sort the doc for frequency: bow_doc
bow_doc = sorted(doc, key=lambda w: w[1], reverse=True)
```

```python
# Print the top 5 words of the document alongside the count
for word_id, word_count in bow_doc[:10]:
    print(dct.get(word_id), word_count)
```

>**TF-IDF with gensim**

```python
from gensim.models.tfidfmodel import TfidfModel
```

```python
print(type(corpus), len(corpus))
print(type(doc), len(doc))
```

```python
print(doc)
```

```python
# Create a new TfidfModel using the corpus: tfidf
tfidf = TfidfModel(corpus, smartirs='ntc')
```

```python
print(type(tfidf))
```

```python
# Calculate the tfidf weights of doc: tfidf_weights
tfidf_weights = tfidf[doc]
```

```python
print(type(tfidf_weights), len(tfidf_weights))
```

```python
# Print the first five weights
print(tfidf_weights)
```

```python
# Sort the weights from highest to lowest: sorted_tfidf_weights
sorted_tfidf_weights = sorted(tfidf_weights, key=lambda w: w[1], reverse=True)
```

```python
sorted_tfidf_weights
```

```python
# Print the top 5 weighted words
for term_id, weight in sorted_tfidf_weights[:5]:
    print(dct.get(term_id), weight)
```
