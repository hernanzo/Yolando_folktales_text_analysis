```python
with open('don_juan_arcarpe_vol1_4.txt', encoding='utf-8-sig') as yolando_1_4: 
    yol_1 = yolando_1_4.read()
with open('el_cuerpo_sin_alma_2_vol1_7.txt', encoding='utf-8-sig') as yolando_1_7: 
    yol_2 = yolando_1_7.read()
```

```python
#filtering text, removing weird characters
fil_yol_1 = re.sub('[^\\w\\s]', '', yol_1)
print(type(fil_yol_1), len(fil_yol_1))
#filtering text, removing weird characters
fil_yol_2 = re.sub('[^\\w\\s]', '', yol_2)
print(type(fil_yol_2), len(fil_yol_2))
```

```python
#lower case the string
lower_yol_1 = str.lower(fil_yol_1)
print(type(lower_yol_1), len(lower_yol_1))
#lower case the string
lower_yol_2 = str.lower(fil_yol_2)
print(type(lower_yol_2), len(lower_yol_2))
```

```python
#create a pattern to remove digits
pattern = r"\d+"
# Match all digits in the string and replace them with an empty string
lower_yol_1 = re.sub(pattern, '', lower_yol_1)
print(len(lower_yol_1), type(lower_yol_1))
lower_yol_2 = re.sub(pattern, '', lower_yol_2)
print(len(lower_yol_2), type(lower_yol_2))
```

```python
nltk.download('punkt')
```

```python
#tokenizing texts into words
tokenized_yol1 = nltk.word_tokenize(lower_yol_1)
print(type(tokenized_yol1))
print(len(tokenized_yol1))

tokenized_yol2 = nltk.word_tokenize(lower_yol_2)
print(type(tokenized_yol2))
print(len(tokenized_yol2))
```

```python
corpus_yol = []
corpus_yol.append(tokenized_yol1)
corpus_yol.append(tokenized_yol2)
```
